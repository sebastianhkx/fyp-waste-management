```python
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
```

```python
default_data = pd.read_csv("data.csv")
```

```python
default_data.rename(columns={'Est. Average Daily Occupancy (Continuous Basis)':'Daily_Occupancy'},
inplace=True)
```

```python
default_data
```

```python
def convertint(x):
    x=int(x.replace(',', ''))
    return x
```

```python
default_data['Waste_amount'] = default_data['audit_avg'].apply(convertint)
```

```python
default_data['GFA'] = default_data['GFA'].apply(convertint)
```

```python
default_data['Daily_Occupancy'] = default_data['Daily_Occupancy'].apply(convertint)
```

```python
default_data
```

```python
def typegroup(x):
    if x == "Residential":
        y=2
    elif x == "Commercial":
        y=1
    elif x == "Industrial":
        y=5
    elif x == "Institutional":
        y=4
    elif x == "Educational":
        y=3
    elif x == "Storage":
        y=6
    return y
```

```python
default_data['Typology Group'] = default_data['Typology'].apply(typegroup)
```

```python
default_data
```

```python
def subtypegroup(x):
    if x =='Apartment':
        y=5
    elif x =='Landed Property':
        y=4
    elif x =='Retail Mixed':
        y=1
    elif x =='Retail Non-Food':
        y=3
    elif x =='Office':
        y=2
    elif x =='Data Center':
        y=13
    elif x =='Manufacturing':
        y=12
    elif x =='Airport':
        y=9
    elif x =='Police Station':
        y=10
    elif x =='Fire Station':
        y=11
    elif x =='Nurseries':
        y=7
    elif x =='University':
        y=6
    elif x =='Student residences':
        y=8
    elif x =='Warehouse':
        y=14
    elif x =='Garage':
        y=15
    return y
```

```python
default_data['Sub-Type Group'] = default_data['Sub-Type'].apply(subtypegroup)
```

```python
default_data
```

```python
def wastegroup(x):
    if x<500:
        y="Lowest Waste Production (<500KG)"
    elif x>=500 and x<1000:
        y="Low Waste Production (500-1000KG)"
    elif x>=1000 and x<1500:
        y="Moderate Waste Production (1000-1500KG)"
    elif x>=1500 and x<2000:
        y="Moderately High Waste Production (1500-2000KG)"
    elif x>=2000 and x<2500:
        y="High Waste Production (2000-2500KG)"
    elif x>=2500 and x<3000:
        y="Higher Waste Production (2500-3000KG)"
    elif x>=3000:
        y="Highest Waste Production (>3500KG)"
    return y
```

```python
default_data['Waste Production Category'] = default_data['Waste_amount'].apply(wastegroup)
```

```python
default_data
```

```python
# Import essential models and functions from sklearn
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

# Extract Response and Predictors
predictors = ["Typology Group", "Sub-Type Group", "GFA", "Daily_Occupancy"]

y = pd.DataFrame(default_data['Waste Production Category'])
X = pd.DataFrame(default_data[predictors]) 

# Split the Dataset into Train and Test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)

# Decision Tree using Train Data
dectree = DecisionTreeClassifier(max_depth = 3)  # create the decision tree object
dectree.fit(X_train, y_train)                    # train the decision tree model

# Predict Response corresponding to Predictors
y_train_pred = dectree.predict(X_train)
```

```python
import pickle 
pickle_out = open("classifier2.pkl", mode = "wb") 
pickle.dump(dectree, pickle_out) 
pickle_out.close()
```
